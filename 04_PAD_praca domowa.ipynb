```python
import pandas as pd
import numpy as np
```

```python
df_orders = pd.read_csv('D:/Studia/St II Sem 1/PAD/4/orders.csv')
df_orders.head(20)
```

```python
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'])
df_orders['order_date'][0]
```

```python
df_orders.info()
```

```python
df_orders['tshirt_category'].value_counts()
```

```python
# nie łatwiej to zrobić tak:
# (albo inaczej, czy to nie jest best practice skoro istnieje metoda stworzona dokładnie do zamiany str)
df_orders['tshirt_category'] = df_orders['tshirt_category'].str.replace('Wh Tshirt','White T-Shirt')
df_orders['tshirt_category'] = df_orders['tshirt_category'].str.replace('Bl Tshirt','Black T-Shirt')
df_orders['tshirt_category'].value_counts()
```

```python
df_orders['tshirt_type'] = df_orders['tshirt_category'].apply(lambda x: x if len(x.split()) < 3 else 'T-shirt')
```

```python
# nie jest napisane co ma się dziać dla bluz i koszulek tenisowych
df_orders['tshirt_size'] = df_orders['tshirt_category'].apply(lambda x: x[-1] if len(x.split()) > 2 else np.nan)
```

```python
# też nie jest napisane co ma być jeśli chodzi o kolor dla powyższych
df_orders['tshirt_colour'] = df_orders['tshirt_category'].apply(lambda x: x.split()[0] if len(x.split()) > 2 else np.nan)
```
